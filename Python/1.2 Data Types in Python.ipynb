# Data Types in Python


# Numeric Data Types (Anything related to numbers)

```python
age = 30 # Age is a Numeric Integer Data Type
sal = 20002.2 # sal is a float(decimal) Data Type
comp = 2+3j # comp is a Complex Data where i and j are imaginary values
```

# String Data Type (Anything related to Character (Alphabet and Special Char)

```python
name = 'Vikram'
address = '# 65 ABC Street, XYZ City Pin 123456 !'
a = 'y'
```

# Boolean Data Type (Check a condition)

```python
price = 34.5
check = True
print(price)
print(check)
```

# Check the Type of data Type with type() command/function


```python
print(type(name))
print(type(address))
print(type(price))
print(type(check))
```

# Data Type Conversions

# Implicit Conversion


```python
income = 2000
extra_income = 200.50
total_income = income+extra_income # Here total income had a float data type assigned by the interpreter itself
print(total_income)
print(type(total_income))
```

# Explicit Conversion

```python
age = 12
tax = '18'
total = age+tax
```

```python
age = 12
tax = '18'
total = age+int(tax)
print(total)
```

# Typecasting, or type conversion, is a method of changing an entity from one data type to another

```python
income = 9000.00
print(int(income))
```

```python
cost = 56.87
str(cost)  # Note the type of stry is now a string
```

```python
a,b=0,10
print(bool(a))
print(bool(b))
```
