# Relational Operators

```python
x_age = 45
y_age = 54
print(x_age > y_age)
print(x_age == y_age)
print(x_age < y_age)
print(x_age <= y_age)
print(x_age >= y_age)
print(x_age != y_age)
```

# Logical Operators

```python
x = True
y = False
print(x > y) # Check if x is greater than y and print result
print(x or y) #  Check if x is True or Y is True and print result
print(not x) # x is true so not of x is False
print(not y) # y is false and hence not of y is true
```

#    Membership operator

```python
string = 'Hello'
var = 'o'
var1 = 'g'
print(var in string)
print(var1 in string)
```

```python
string = 'Hello'
var = 'o'
var1 = 'g'
print(var not in string)
print(var1 not in string)
```

# Bit Wise Operators


```python
value = 10
(value % 3 == 0) & (value % 2 == 0)
```

```python
value = 10
(value % 3 == 0) | (value % 2 == 0)
```
